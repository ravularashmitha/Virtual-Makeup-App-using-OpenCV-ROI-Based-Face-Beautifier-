```python
# with croping and selection img from device


import cv2
import numpy as np
from tkinter import Tk
from tkinter.filedialog import askopenfilename

# -------------------------------
# Step 1: Select image via dialog
# -------------------------------
file_path = askopenfilename(
    title="Select an image",
    filetypes=[("Image Files", "*.jpg *.jpeg *.png *.bmp *.tif *.tiff")]
)

# If user cancels file selection, exit program
if not file_path:
   #"No file selected. Exiting..."
    exit()

# -------------------------------
# Step 2: Load image and resize
# -------------------------------
img = cv2.imread(file_path)         # Load image using OpenCV
img = cv2.resize(img, (300, 400))   # Resize to manageable size
final = img.copy()                  # Copy to keep original unchanged for editing

# -------------------------------
# Step 3: Global variables
# -------------------------------
x1 = y1 = x2 = y2 = None  # Coordinates for cropping rectangle
drawing = False            # Flag to indicate if mouse is drawing rectangle 
crop_done = False          # Flag to indicate if a crop is completed
mode = None                # Current mode: 'face' or 'lips'
mask = None                # Mask for selected HSV range
crop = []                  # Stores crop coordinates

# -------------------------------
# Step 4: Mouse callback for cropping
# -------------------------------
def mouse_crop(event, x, y, flags, param):
    """
    Handles mouse events to draw a rectangle and store crop coordinates
    """
    global x1, y1, x2, y2, drawing, crop_done, crop

    if event == cv2.EVENT_LBUTTONDOWN:  # Left mouse button pressed
        drawing = True                   # Start drawing -- so that other drawing is not happening 
        x1, y1 = x, y                     # Store starting coordinates

#    elif event == cv2.EVENT_MOUSEMOVE and drawing:  # Mouse is moving while pressed
#        temp = final.copy()               # Make temporary copy to show live rectangle
#        cv2.rectangle(temp, (x1, y1), (x, y), (255, 255, 255), 2)  # Draw rectangle
#        cv2.imshow("makeup", temp)       # Show live rectangle

    elif event == cv2.EVENT_LBUTTONUP:   # Left mouse button released
        drawing = False                   # Stop drawing
        x2, y2 = x, y                     # Store ending coordinates
        crop_done = True                  # Crop is completed
        # Draw rectangle on temporary display only
        temp = final.copy()
        cv2.rectangle(temp, (x1, y1), (x2, y2), (255, 255, 255), 2)
        cv2.imshow("makeup", temp)
        crop = [y1, y2, x1, x2]          # Store crop coordinates

# -------------------------------
# Step 5: Trackbar callback (does nothing)
# -------------------------------
def nothing(a):
    pass  # Required placeholder for trackbars

# -------------------------------
# Step 6: Create OpenCV window and trackbars
# -------------------------------
cv2.namedWindow("makeup", cv2.WINDOW_AUTOSIZE)  # Window for displaying image
cv2.setMouseCallback("makeup", mouse_crop)      # Assign mouse callback

# HSV range trackbars
cv2.createTrackbar("hmin", "makeup", 0, 179, nothing)
cv2.createTrackbar("hmax", "makeup", 179, 179, nothing)
cv2.createTrackbar("smin", "makeup", 0, 255, nothing)
cv2.createTrackbar("smax", "makeup", 255, 255, nothing)
cv2.createTrackbar("vmin", "makeup", 0, 255, nothing)
cv2.createTrackbar("vmax", "makeup", 255, 255, nothing)

# Brightness/Darkness trackbars
cv2.createTrackbar("Face Brightness", "makeup", 0, 100, nothing)
cv2.createTrackbar("Lips Darkness", "makeup", 0, 100, nothing)

# -------------------------------
# Step 7: Main loop
# -------------------------------
while True:
    display = final.copy()  # Make copy to show live edits

    # If crop is done, process selected area
    if crop_done:
        y1, y2, x1, x2 = crop              # Get crop coordinates
        crop_image = display[y1:y2, x1:x2] # Extract cropped image
        hsv = cv2.cvtColor(crop_image, cv2.COLOR_BGR2HSV)  # Convert crop to HSV

        # -------------------------------
        # Step 7a: Read trackbar values
        # -------------------------------
        hm = cv2.getTrackbarPos("hmin", "makeup")
        hx = cv2.getTrackbarPos("hmax", "makeup")
        sm = cv2.getTrackbarPos("smin", "makeup")
        sx = cv2.getTrackbarPos("smax", "makeup")
        vm = cv2.getTrackbarPos("vmin", "makeup")
        vx = cv2.getTrackbarPos("vmax", "makeup")
        face_bri = cv2.getTrackbarPos("Face Brightness", "makeup")
        lips_dar = cv2.getTrackbarPos("Lips Darkness", "makeup")

        # Create HSV mask for selected range
        lower = np.array([hm, sm, vm])
        upper = np.array([hx, sx, vx])
        mask = cv2.inRange(hsv, lower, upper)

        # -------------------------------
        # Step 7b: Clean mask using morphological operations
        # -------------------------------
        kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (5,5))  # Elliptical kernel
        mask = cv2.morphologyEx(mask, cv2.MORPH_CLOSE, kernel)        # Fill small holes
        mask = cv2.morphologyEx(mask, cv2.MORPH_OPEN, kernel)         # Remove small noise

        # -------------------------------
        # Step 7c: Apply brightness/darkness
        # -------------------------------
        preview = crop_image.astype(np.float32)  # Convert to float for safe arithmetic
        if mode == "face":
            preview += face_bri                 # Increase brightness
        elif mode == "lips":
            preview -= lips_dar                 # Decrease brightness
        preview = np.clip(preview, 0, 255).astype(np.uint8)  # Clip values to 0-255

        # Apply mask to keep changes only in selected region
        masked = cv2.bitwise_and(preview, preview, mask=mask)
        mask_inv = cv2.bitwise_not(mask)
        bg = cv2.bitwise_and(crop_image, crop_image, mask=mask_inv)  # Background remains unchanged
        final_crop = cv2.add(bg, masked)  # Combine masked area with background

        # Put edited crop back into the full image
        display[y1:y2, x1:x2] = final_crop

        # Show mask and live preview
        cv2.imshow("Mask (Selected Area)", mask)
        cv2.imshow("Preview", display)
    else:
        # If no crop yet, just show original
        cv2.imshow("makeup", display)

    # -------------------------------
    # Step 8: Key controls
    # -------------------------------
    key = cv2.waitKey(1) & 0xFF

    if key == ord('f'):       # Face mode (increase brightness)
        mode = "face"

    elif key == ord('l'):     # Lips mode (decrease brightness)
        mode = "lips"

    elif key == ord('a') and crop_done:  
        # Apply changes permanently to final image
        final[y1:y2, x1:x2] = display[y1:y2, x1:x2]
        crop_done = False  # Allow new crop

    elif key == ord('s'):     # Save final result to file
        x1 = y1 = x2 = y2 = None
        cv2.imwrite("C:\\Users\\Sana\\Pictures\\Saved Pictures\\makeup_result.jpg", final)
        cv2.putText(final, "Saved!", (50, 50),cv2.FONT_HERSHEY_SIMPLEX, 1,(0, 255, 0) , 2, cv2.LINE_AA)

    elif key == ord('r'):     # Reset crop, mask, trackbars, previews
        x1 = y1 = x2 = y2 = None
        crop_done = False
        crop = []
        mask = None
        display = final.copy()
    
        # Reset trackbars
        cv2.setTrackbarPos("hmin", "makeup", 0)
        cv2.setTrackbarPos("hmax", "makeup", 179)
        cv2.setTrackbarPos("smin", "makeup", 0)
        cv2.setTrackbarPos("smax", "makeup", 255)
        cv2.setTrackbarPos("vmin", "makeup", 0)
        cv2.setTrackbarPos("vmax", "makeup", 255)
        cv2.setTrackbarPos("Face Brightness", "makeup", 0)
        cv2.setTrackbarPos("Lips Darkness", "makeup", 0)

        # Close any preview windows
        cv2.destroyWindow("Mask (Selected Area)")
        cv2.destroyWindow("Preview")

    elif key == ord('q'):     # Quit program
        break

# -------------------------------
# Step 9: Cleanup
# -------------------------------
cv2.destroyAllWindows()  # Close all OpenCV windows
```
